```python
import time
import numpy as np

import research_functions_vectorised
import research_functions
```

```python
#  

are_multisets_equal_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(0, 2, n)
    y = np.random.randint(0, 2, n)
    are_multisets_equal_data += [(x, y)]

max_prod_mod_3_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(1, 100, n)
    max_prod_mod_3_data += [x]

convert_image_data = []
for n in [10, 100, 1000]:
    image = np.random.random((n, n, 10))
    weights = np.random.random(10)
    convert_image_data += [(image, weights)]
print(len(convert_image_data))

rle_scalar_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(1, 3, (n, 2))
    y = np.random.randint(1, 3, (n, 2))
    shapes = np.random.randint(1, 3, n)
    x[:, 1] = shapes
    y[:, 1] = shapes

    rle_scalar_data += [(x, y)]

cosine_distance_data = []
for n in [10, 100]:
    X = np.random.random((n, n))
    Y = np.random.random((n, n))
    cosine_distance_data += [(X, Y)]
```

```python
#     

def count_time(func, *data):
    tick = time.perf_counter()
    func(*data)
    return time.perf_counter() - tick
```

```python
#    

#are_multisets_equal_data
are_multisets_equal_time = []
are_multisets_equal_vectorized_time = []

for x in are_multisets_equal_data:
    are_multisets_equal_time += [count_time(research_functions.are_multisets_equal, x[0].tolist(), x[1].tolist())]
    are_multisets_equal_vectorized_time += [count_time(research_functions_vectorised.are_multisets_equal, x[0], x[1])]
# max_prod_mod_3
max_prod_mod_3_time = []
max_prod_mod_3_vectorised_time = []

for x in max_prod_mod_3_data:
   max_prod_mod_3_time += [count_time(research_functions.max_prod_mod_3, x.tolist())]
   max_prod_mod_3_vectorised_time += [count_time(research_functions_vectorised.max_prod_mod_3, x)]

# convert_image

convert_image_data_time = []
convert_image_data_vectorised_time = []
for x in convert_image_data:
   convert_image_data_time += [count_time(research_functions.convert_image, x[0].tolist(),  x[1].tolist())]
   convert_image_data_vectorised_time += [count_time(research_functions_vectorised.convert_image, x[0], x[1])]


# rle_scalar
rle_scalar_time = []
rle_scalar_vectorized_time = []
for x in rle_scalar_data:
   rle_scalar_time += [count_time(research_functions.rle_scalar, x[0].tolist(), x[1].tolist())]
   rle_scalar_vectorized_time += [count_time(research_functions_vectorised.rle_scalar, x[0], x[1])]

# cosine_distance
cosine_distance_time = []
cosine_distance_vectorized_time = []
for x in cosine_distance_data:
    cosine_distance_time += [count_time(research_functions.cosine_distance, x[0].tolist(), x[1].tolist())]
    cosine_distance_vectorized_time += [count_time(research_functions_vectorised.cosine_distance, x[0], x[1])]
    

```

```python
plt.figure(figsize=(10, 15))

plt.subplot(3, 2, 1)
plt.title("  are_multisets_equal")
plt.xlabel(' ')
plt.ylabel(' ()')

plt.grid(True)
plt.plot([10, 100, 1000, 10000], are_multisets_equal_vectorized_time, marker = '.', color = 'r', label = ' ')
plt.plot([10, 100, 1000, 10000], are_multisets_equal_time, marker = '.', color = 'b', label = ' ')
plt.xscale('log')
plt.legend()


plt.subplot(3, 2, 2)
plt.title("  max_prod_mod_3")
plt.xlabel(' ')
plt.ylabel(' ()')

plt.grid(True)
plt.plot([10, 100, 1000, 10000], max_prod_mod_3_vectorised_time, marker = '.', color = 'r', label = ' ')
plt.plot([10, 100, 1000, 10000], max_prod_mod_3_time, marker = '.', color = 'b', label = ' ')
plt.xscale('log')
plt.legend()


plt.subplot(3, 2, 3)
plt.title("  convert_image")
plt.xlabel(' ')
plt.ylabel(' ()')

plt.grid(True)
plt.plot([10, 100, 1000], convert_image_data_vectorised_time, marker = '.', color = 'r', label = ' ')
plt.plot([10, 100, 1000], convert_image_data_time, marker = '.', color = 'b', label = ' ')
plt.xscale('log')
plt.legend()


plt.subplot(3, 2, 4)
plt.title("  rle_scalar")
plt.xlabel(' ')
plt.ylabel(' ()')

plt.grid(True)
plt.plot([10, 100, 1000, 10000], rle_scalar_vectorized_time, marker = '.', color = 'r', label = ' ')
plt.plot([10, 100, 1000, 10000], rle_scalar_time, marker = '.', color = 'b', label = ' ')
plt.xscale('log')
plt.legend()


plt.subplot(3, 2, 5)
plt.title("  cosine_distance")
plt.xlabel(' ')
plt.ylabel(' ()')

plt.grid(True)
plt.plot([10, 100], cosine_distance_vectorized_time, marker = '.', color = 'r', label = ' ')
plt.plot([10, 100], cosine_distance_time, marker = '.', color = 'b', label = ' ')
plt.xscale('log')
plt.legend()

plt.show()
```
