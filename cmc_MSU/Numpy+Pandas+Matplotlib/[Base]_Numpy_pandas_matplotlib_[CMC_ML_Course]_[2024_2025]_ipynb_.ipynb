```python
import matplotlib.pyplot as plt #     


df.publishedAt = pd.to_datetime(df.publishedAt)
_df = df.drop_duplicates(['title', 'publishedAt'], keep = 'first')[['publishedAt', 'title']]
_df.iloc[:, 0] = _df.iloc[:, 0].dt.date
_df = _df.groupby('publishedAt').size().reset_index()

plt.figure(figsize = (9, 5))
plt.title('      ')
plt.grid(True)
plt.plot( _df['publishedAt'], _df[0], marker = '.', linestyle = '-', color = 'r')

plt.xlabel('')
plt.xticks(rotation = 45)

plt.ylabel(' ')


plt.show()
```

### <font color='DarkOrange'>** 2 [0.5 a]**</font>

  `(bar plot)` -      .

```python
_df = df.groupby('country').size().reset_index()

plt.figure(figsize = (9, 5))
plt.title('     ')

plt.bar(_df['country'], _df[0], color = ['#191970', '#607b8b', '#008b8b', '#104e8b'], width = 0.9)
plt.ylabel(' ')
plt.xlabel('')
plt.show()



```

### <font color='DarkOrange'>** 3 [1 ]**</font>

```python
_df = df[(df['comments_disabled'] == False) & (df['country'] == 'FR')][['view_count', 'comment_count']]

plt.figure(figsize = (9, 5))
plt.title('       ')
plt.xlabel(' ')
plt.ylabel(' ')
plt.grid(True)

plt.scatter(_df['view_count'], _df['comment_count'], alpha = 0.5, color = 'blue')
plt.xscale('log')
plt.yscale('log')
```
